```python
#Experiment2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.cluster import KMeans
```

```python
df=np.array([[5,3],
         [10,15],
         [24,10],
         [30,45],
         [85,70],
         [71,80],
         [60,78],
         [55,52],
         [80,91]])
```

```python
df
```

```python
plt.scatter(df[:,0],df[:,1],label='True Position',color='purple')
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(df)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
plt.scatter(df[:,0],df[:,1],c=kmeans.labels_,cmap='BrBG')
```

```python
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black')
```

```python
plt.scatter(df[:,0],df[:,1],c=kmeans.labels_,cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black')
```

```python
#clustering with cluster data set
plt.figure(figsize=(10,6))
plt.scatter(df1['Feature 1'],df1['Feature 2'],label='True Position',color='purple',s=5)
```

```python
kmeans_df1 = KMeans(n_clusters=3, random_state=0, n_init=10) # Assuming 3 clusters
clusters_df1 = kmeans_df1.fit_predict(df1[['Feature 1', 'Feature 2']])
plt.figure(figsize=(10, 8))
plt.scatter(df1['Feature 1'], df1['Feature 2'], c=clusters_df1, cmap='BuPu', s=10)
plt.scatter(kmeans_df1.cluster_centers_[:, 0], kmeans_df1.cluster_centers_[:, 1], s=20, c='red', marker='X', label='Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
